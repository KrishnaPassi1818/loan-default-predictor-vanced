```python
import pandas as pd
```

```python
# 1. Load the dataset
df = pd.read_csv('../dataset/Loan_Default.csv')
```

```python
# 2. Strip leading/trailing whitespaces from column names & text values
df.columns = df.columns.str.strip()
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()
```

```python
# 3. Drop uninformative / zero-variance / non-predictive columns
cols_to_drop = ['ID', 'year', 'construction_type', 'Secured_by', 'Security_Type']
df = df.drop(columns=[col for col in cols_to_drop if col in df.columns])
```

```python
# 4. Separate target classes
df_paid = df[df['Status'] == 0]
df_default = df[df['Status'] == 1]
```

```python
# 5. Downsample majority class to match minority class
df_paid_downsampled = df_paid.sample(n=len(df_default), random_state=42)
```

```python
# 6. Combine and shuffle rows
df_balanced_clean = pd.concat([df_paid_downsampled, df_default]) \
                      .sample(frac=1, random_state=42) \
                      .reset_index(drop=True)
```

```python
# 7. Print summary and export
print("Final Dataset Shape:", df_balanced_clean.shape)
print("\nClass Distribution:")
print(df_balanced_clean['Status'].value_counts())
```

```python
df_balanced_clean.to_csv('../dataset/Loan_Default_Clean_Balanced_5050.csv', index=False)
print("\nSuccessfully exported 'Loan_Default_Clean_Balanced_5050.csv'!")
```
